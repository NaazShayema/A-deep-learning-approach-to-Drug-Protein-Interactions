```r
# Don't run this (change runtime to python)

def read_fasta(filename):
    sequences = []
    with open(filename, 'r') as file:
        current_id = None
        sequence = ''
        for line in file:
            #line = line.strip()
            #print(line)
            if line.startswith('>'):
                if current_id:
                    sequences.append((current_id, sequence))
                current_id = line[1:]
                sequence = ''
            else:
                #sequence += line
                sequence += line

        if current_id:
            sequences.append((current_id, sequence))
    return sequences

def write_fasta(sequence_id, sequence, output_filename):
    with open(output_filename, 'w') as file:
        file.write(f">")
        #print("output:" + sequence)
        file.write(sequence)
        file.write(f"\n")
```

```r
input_filename="/content/drive/MyDrive/Datasets/gene.fasta"
sequences = read_fasta(input_filename)
```

```r
import shutil

result_folder = '/content/drive/MyDrive/Datasets/GENE_PROCESSED_FILE'

for index, (sequence_id, sequence) in enumerate(sequences):
    output_filename = f"{sequence_id.split('|')[1].split(' ')[0]}.fasta"
    write_fasta(sequence_id, sequence, output_filename)
    print(f"Writing {index + 1}-th file")
    shutil.copyfile(output_filename, f"{result_folder}/{output_filename}")
    print(f"Written {index + 1}-th file")
```
