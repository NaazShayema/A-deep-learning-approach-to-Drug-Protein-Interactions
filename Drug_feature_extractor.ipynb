**Creating SMILES csv file for Drug_Ids**

```python
### matched drug Ids with SMILE sequence


# Load necessary libraries
library(dplyr)

# Read the CSV files
all_data <- read.csv("/content/drive/MyDrive/Datasets/all_1.csv")  ## seq
uniprot_data <- read.csv("/content/drive/MyDrive/Datasets/Drug_IDs.csv") # id

# Merge the data based on 'ID' column
merged_data <- inner_join(all_data, uniprot_data , by = "DrugBank.ID")

# Filter rows where 'SMILES' is not empty and not null
smiles_df <- merged_data[!is.na(merged_data$SMILES) & merged_data$SMILES !="", ]

# Write the resulting dataframe to a CSV file
write.csv(smiles_df[, c("DrugBank.ID", "SMILES")], 'smiles_output_2.csv', row.names=FALSE)
```

```python
library(dplyr)

# Read the CSV files
smiles_output <- read.csv('/content/smiles_output_2.csv')
drug_ids <- read.csv('/content/drive/MyDrive/Datasets/Drug_IDs_3.csv')

# Check if all values in 'Drugbank.ID' of smiles_output are present in 'Drugbank.ID' of drug_ids
all_present <- all(smiles_output$Drugbank.ID %in% drug_ids$Drugbank.ID)

# Print the result
if (all_present) {
  cat("All values in 'Drugbank.ID' of smiles_output are present in 'Drugbank.ID' of drug_ids.")
} else {
  cat("Not all values in 'Drugbank.ID' of smiles_output are present in 'Drugbank.ID' of drug_ids.")
}
```

```python
#### matched drug IDs with no SMILE sequence


# Load necessary libraries
library(dplyr)

# Read the CSV files
all_data <- read.csv("/content/drive/MyDrive/Datasets/all_1.csv")
uniprot_data <- read.csv("/content/drive/MyDrive/Datasets/Drug_IDs.csv")

# Merge the data based on 'ID' column
merged_data <- inner_join(all_data, uniprot_data , by = "DrugBank.ID") %>%
              filter(is.na(SMILES) | SMILES == "")


# Write the resulting dataframe to a CSV file
write.csv(merged_data[, c("DrugBank.ID")], 'matched_ids_and_noSmile.csv',row.names=FALSE)
```

```python
### unmatched drug IDs with SMILE sequence


# Load necessary libraries
library(dplyr)

# Read the CSV files
all_data <- read.csv("/content/drive/MyDrive/Datasets/all_1.csv")
uniprot_data <- read.csv("/content/drive/MyDrive/Datasets/Drug_IDs.csv")

# Create a new CSV file with only 'ID' column for unmatched IDs
unmatched_ids <- anti_join(uniprot_data, all_data, by = "DrugBank.ID")
write.csv(unmatched_ids, "unmatched_ids.csv", row.names=FALSE)
```
