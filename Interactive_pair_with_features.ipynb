**Creating 'Interactive_pair_with_features.csv'**

which contains drug & protein pair which interact with each other and their corresponding features

```python
import pandas as pd

# Read Interactive.csv file
interactive_df = pd.read_csv('/content/drive/MyDrive/Datasets/D_P_Interaction/Interactive.csv')

# Initialize an empty DataFrame to store the merged data
merged_df = interactive_df.copy()

# List of paths to other CSV files
#other_uniprot_files = ['/content/drive/MyDrive/Datasets/Protein_Features/AAC_updated_final.csv','/content/drive/MyDrive/Datasets/Protein_Features/DC_updated_final.csv','/content/drive/MyDrive/Datasets/Protein_Features/CTDC_updated_fianl.csv','/content/drive/MyDrive/Datasets/Protein_Features/CTDD_updated_final.csv','/content/drive/MyDrive/Datasets/Protein_Features/CTDT_updated_final.csv']
other_drugbank_files = ['/content/drive/MyDrive/Datasets/Drug_Features/DrugAminoAcidCount_final.csv','/content/drive/MyDrive/Datasets/Drug_Features/DrugAtomCount_final.csv','/content/drive/MyDrive/Datasets/Drug_Features/DrugAutocorrelationMass_final.csv','/content/drive/MyDrive/Datasets/Drug_Features/DrugCarbonTypes_final.csv','/content/drive/MyDrive/Datasets/Drug_Features/DrugKierHallSmarts_final.csv']
other_uniprot_files = ['/content/drive/MyDrive/Datasets/Protein_Features/AAC_updated_final.csv','/content/drive/MyDrive/Datasets/Protein_Features/DC_updated_final.csv','/content/drive/MyDrive/Datasets/Protein_Features/CTDC_updated_fianl.csv','/content/drive/MyDrive/Datasets/Protein_Features/CTDD_updated_final.csv','/content/drive/MyDrive/Datasets/Protein_Features/CTDT_updated_final.csv']

# # Merge data from files with 'UniProt ID' column
# for file in other_uniprot_files:
#     temp_df = pd.read_csv(file)
#     merged_df = pd.merge(merged_df, temp_df, on='UniProt ID', how='inner')

# Merge data from files with 'Drugbank ID' column
for file in other_drugbank_files:
    temp_df = pd.read_csv(file)
    merged_df = pd.merge(merged_df, temp_df, on='DrugBank ID', how='inner')

# Merge data from files with 'UniProt ID' column
for file in other_uniprot_files:
    temp_df = pd.read_csv(file)
    merged_df = pd.merge(merged_df, temp_df, on='UniProt ID', how='inner')


# Write the merged data to a new CSV file
merged_df.to_csv('Interactive_pair_with_features.csv', index=False)
```
