**creating non_interactive file with drug and protein ids**

```python
import pandas as pd
```

```python
prot_Id=pd.read_csv("/content/drive/MyDrive/Datasets/D_P_Interaction/Unique_prot_Ids.csv")
prot_Id

drug_Id=pd.read_csv("/content/drive/MyDrive/Datasets/D_P_Interaction/unique_Drug_IDs.csv")
drug_Id

interaction=pd.read_csv("/content/drive/MyDrive/Datasets/D_P_Interaction/Interactive.csv")
interaction
```

```python
random_drug_id = drug_Id['DrugBank ID'].sample().values[0]
random_drug_id
```

```python
random_prot_id = prot_Id['UniProt ID'].sample().values[0]
random_prot_id
```

```python
interaction.head()
```

```python
import pandas as pd

count=0
ni_df=pd.DataFrame(columns=['DrugBank_ID','UniProt_ID'])

while count<=20000:
  random_drug_id = drug_Id['DrugBank ID'].sample().values[0]
  random_prot_id = prot_Id['UniProt ID'].sample().values[0]
  indices = row_index = interaction.loc[(interaction['DrugBank ID'] == random_drug_id) & (interaction['UniProt ID'] == random_prot_id)].index
  if len(indices) > 0:
      row_index = indices[0]
      selected_row = interaction.iloc[row_index]
      print(selected_row)
  else:
    count+=1
    #ni_df=ni_df.append({'DrugBank_ID':random_drug_id,'UniProt_ID':random_prot_id},ignore_index=True)
    ni_df.loc[len(ni_df)] = {'DrugBank_ID':random_drug_id,'UniProt_ID':random_prot_id}
    #print(f"{DrugBank ID} non-interacting with {UniProt ID}")
    print("non interactive")

```

```python
print(ni_df)
```
